Question 21
Repeat step 19 and 20 for other combinations of independent variables wrt price. Observations? Inferences?

```
redundant = []
for i in df.describe(include=np.object).columns:
    redundant.append(i)
    for j in df.describe(include=np.object).columns:
      if i==j or j in redundant:
        continue
      else:
        temp = df.groupby([i,j])['price'].mean()
        print(temp, end='\n\n')
        temp = df[[i,j,'price']]
        for_pivot = temp.groupby([i,j], as_index=False).mean()
        piv = for_pivot.pivot(index = i, columns = j)
        piv.replace(np.nan, '-', inplace= True)
        print('Pivot Table')
        print(piv)
        print('------', end='\n\n')
```

Question 24
Perform one way ANOVA test using f_oneway() of stats to check if different groups of “drive-wheels” are correlated with “price”.

```
grp2=df[['drive-wheels', 'price']].groupby(['drive-wheels'])
grp2.head()
```

```
f_val, p_val = stats.f_oneway(grp2.get_group('fwd')['price'], grp2.get_group('rwd')['price'], grp2.get_group('4wd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val)
```
